```python
import numpy as np
import random as rnd
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default='iframe'

Lx=Ly=25.4
n_plot=100
N_sim=20
d=-25.4

x_sup=np.linspace(0,25.4,n_plot)
y_sup=np.linspace(0,25.4,n_plot)
z_sup=np.zeros([n_plot,n_plot])
z_sup2=d+z_sup

x_plotf=[]
y_plotf=[]

x_ploterr=[]
y_ploterr=[]

targets=0
colors=[]
aux=np.linspace(d, 0,n_plot)
for i in range(N_sim):
    x=rnd.uniform(0,Lx)
    y=rnd.uniform(0,Ly)
    phi=rnd.uniform(0,2*np.pi)
    theta=np.arccos((1-rnd.random())**(1/4))
    radial=d*np.tan(theta)
    x_final=x+radial*np.cos(phi)
    y_final=y+radial*np.sin(phi)
    xplot=x + aux*np.tan(theta)*np.cos(phi)
    yplot=y + aux*np.tan(theta)*np.sin(phi)
    
    if ((x_final<=Lx) & (x_final>=0)):
        if((y_final<=Ly) & (y_final>=0)):
            colors.append("blue")
        
        else:
            colors.append("red")
    else:
        colors.append("red")
        
    x_plotf.append(xplot)
    y_plotf.append(yplot)

    
    
fig = go.Figure(
    data=[
    go.Surface(x=x_sup, y=y_sup, z=z_sup2, colorscale='viridis'),
    go.Surface(x=x_sup, y=y_sup, z=z_sup),
    go.Surface(x=x_sup, y=y_sup, z=z_sup2, colorscale='viridis'),
    go.Surface(x=x_sup, y=y_sup, z=z_sup)],
    
    layout=go.Layout(
        xaxis=dict(range=[-2, Lx+2], autorange=False, zeroline=False),
        yaxis=dict(range=[-2, Ly+2], autorange=False, zeroline=False),
        updatemenus=[dict(type="buttons",
                          buttons=[dict(label="Play",
                                        method="animate",
                                        args=[None]),
                                   
                                   dict(label="Pause",
                                        method="animate",
                                        args=[[None], {"mode" : "immediate" }])])],
         scene={
        'xaxis': {'range': [-5, Lx], 'rangemode': 'tozero', 'tickmode': "linear", 'tick0': -5, 'dtick': 5},
        'yaxis': {'range': [-5, Lx], 'rangemode': 'tozero', 'tickmode': "linear", 'tick0': -5, 'dtick': 5},
        'zaxis': {'range': [-26, 2], 'rangemode': 'tozero', 'tickmode': "linear", 'tick0': -5, 'dtick': 5},
        'aspectratio': {
            'x': 0,
            'y': 0,
            'z': 0,
        },
        'aspectmode':"cube"
         }),
    
    
    frames=[go.Frame(
        data=[go.Scatter3d(
            x=x_plotf[k],
            y=y_plotf[k],
            z=aux,
            mode="lines",
            line=dict(color=colors[k], width=2))])

        for k in range(N_sim)]
)
fig.show()
```
